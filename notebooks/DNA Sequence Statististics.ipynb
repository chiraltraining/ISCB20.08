```python
def readFASTA(inputfile): 
    """Reads a FASTA file and returns it with special character removed"""
    for record in SeqIO.parse(inputfile, "fasta"): 
        seq = record.seq
    return seq 
```

```python
# reading data 
den1 = readFASTA('../data/den1.fasta')
den2 = readFASTA('../data/den2.fasta')
den3 = readFASTA('../data/den3.fasta')
den4 = readFASTA('../data/den4.fasta')
```

```python
from collections import Counter 
def nt_frequency(seq): 
    """Returns the frequency distribution of nucleotides"""
    freqs = Counter(seq)
    return freqs 
```

```python
def calculate_gc(seq): 
    """Function to calculate GC ratio"""
    no_of_g = seq.count('G')
    no_of_c = seq.count('C')
    total = no_of_c + no_of_g
    gc = round(total/len(seq) * 100, 2)
    return gc 
```

```python
calculate_gc(den1)
```

```python
calculate_gc(den2)
```

```python
calculate_gc(den3)
```

```python
calculate_gc(den4)
```

```python
def calculate_at(seq): 
    """Function to calculate AT ratio"""
    no_of_a = seq.count('A')
    no_of_t = seq.count('T')
    total = no_of_a + no_of_t
    at = round(total/len(seq) * 100, 2)
    return at 
```

```python
calculate_at(den2)
```

```python
calculate_at(den3)
```

```python
calculate_at(den4)
```

## Sliding Window Analysis 

```python
def subseqgc(seq, window=300): 
    """Function to calculate sub-sequence GC ratio"""
    res = [] 
    for i in range(0, len(seq)-window+1, window): 
        subseq = seq[i:i+window] 
        gc = calculate_gc(subseq) 
        res.append(gc)
    
    return res    
```

```python
subseqgc(den3, 400)
```

```python
def subseqat(seq, window=300): 
    """Function to calculate sub-sequence GC ratio"""
    res = [] 
    for i in range(0, len(seq)-window+1, window): 
        subseq = seq[i:i+window] 
        at = calculate_at(subseq) 
        res.append(at)
    
    return res  
```
