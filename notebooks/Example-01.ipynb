# Problem Statement: Find out from what species an unknown DNA sequence came from?

## Step-1: Running BLAST

```python
from Bio.Blast import NCBIWWW
file = open("../data/refsequnknown.fa").read() 
result = NCBIWWW.qblast("blastn", "nt", file)
```

## Step-2: BLAST Record

```python
from Bio.Blast import NCBIXML
blast_record = NCBIXML.read(result)
```

## Step-3: Evaluation

```python
e_value = 0.01 
for alignment in blast_record.alignments: 
    for hsp in alignment.hsps: 
        if hsp.expect < e_value: 
            print("Alignment")
            print("Sequence:", alignment.title)
            print("Length: ", alignment.length)
            print("E-value: ", hsp.expect)
            print(hsp.query)
            print(hsp.match)
            print(hsp.sbjct)
```
